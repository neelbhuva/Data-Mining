##
From the dataset, it is observed that the attributes are a combination of continuous, nominal, ordinal and interval variables. The first step in the analaysis of the Income dataset is to make decision considering the different types of data.

## Scaling and Standardization:
For the continuous variables, we normalized the data using standard deviation to scale the data between 0 to 1. The feature scaling is done to make the assumption that all the features has the equal opportunity to influence the weight, which more really reflects the knowledge you know about the data. 

## Proximity Measures for attributes:
We will discuss about the different similarity or distance measures used for different attributes.

### Continuous attributes: 
The continuous attributes that are in the data set are age, fnlwgt, hour_per_week. We will standadize the data using standard deviation scaling function. We can calculate the distances using Minkowski distance for n = 3 and L2 norm.

Dissimilarity: d = minkowski with norm 2, similairty: s = 1/e^d. As the distance increases, e^d increases and the similarity decreases.

### Ordinal Attributes:

Dissimilarity d = |x-y|/(n-1) and similarity s = 1 - d.

### Nominal Attributes:
The nominal attributes in the data are workclass, education, occupation, marital_status, relationship, race, gender, native_country.

Dissimilarity d = 0 if x = y, or d = 1 if x != y, 
similarity s = 1 if x = y, or s = 0 if x != y.

#### Now we now had a look at all the variables.
#### First, we will consider "age" and "hour_per_week" as continous. Depending on the measures, we can later categorize it or leave it as it is.
#### The attributes "education" and "education_cat" both represent the same thing. As the "education_cat" attribute is an ordinal attribute, it is easy to calculate similarities if it is already mapped or encoded. So we will drop "education".
#### Missing values in "workclass" and "occupation" which is denoted by "?" belong to the same observations. An assumption on this could be made by treating these value as a seperate category. We will name the occupation as "Job_X" and the workclass as "X".
#### Attributes which can be treated as nominal are "workclass", "marital_status", "occupation", "relationship", "race", "gender", "native_country".
#### There are missing values in the attribute "native_country". Approximately 89% of the observations are of the same country i.e. United-States and the missing values are 1.1%. We can replace these missing values with most frequent value as there are only few missing values.

```python
def nom_sim(dframe):
    temp = []
    for i in dframe:
        for j in dframe:
            if i == j:
                a = 1
                temp.append(a)
            else:
                a = 0
                temp.append(a)        
    ar = np.array(temp)
    shape = (len(dframe), len(dframe))
    ar_reshape = ar.reshape(shape,)
    df_sim = pd.DataFrame(data = ar_reshape)
    return(df_sim)


def ord_sim(dframe):
    temp = []
    for i in dframe:
        i = i - 1
        for j in dframe:
            j = j - 1
            d = abs(i - j) / (len(dframe.unique()) - 1)
            s = temp.append(1-d)
    ar = np.array(temp)
    shape = (len(dframe), len(dframe))
    ar_reshape = ar.reshape(shape,)
    df_sim = pd.DataFrame(data = ar_reshape)
    return(df_sim)

def average_sim(df_list):
    average = []
    for i in range(0,520):
        for j in range(0,520):
            avg = 0
            for k in range(0,len(df_list)):
                avg = avg + df_list[k][i][j]
            avg = avg / len(df_list)
            average.append(avg)
    
    ar = np.array(average)
    shape = (520, 520)
    ar_reshape = ar.reshape(shape,)
    df_avg = pd.DataFrame(data = ar_reshape)
    return(df_avg)

df_edu_sim = ord_sim(df["education_cat"])
df_occu_sim = nom_sim(df["occupation"])
df_race_sim = nom_sim(df["race"])
df_country_sim = nom_sim(df["native_country"])
df_marital_sim = nom_sim(df["marital_status"])
#df_marital_sim.head()
df_list = [df_mdist,df_edu_sim,df_occu_sim,df_race_sim,df_country_sim,df_marital_sim]
df_avg = average_sim(df_list)
#df_avg.head()
```
